```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("AirQuality_visualization.csv",delimiter=";")
```

```python
df = df.drop(["Unnamed: 15","Unnamed: 16"], axis=1)
```

```python
df["CO(GT)"] = df["CO(GT)"].str.replace(",",".").astype("float")
df["C6H6(GT)"] = df["C6H6(GT)"].str.replace(",",".").astype("float")
df["T"] = df["T"].str.replace(",",".").astype("float")
df["AH"] = df["AH"].str.replace(",",".").astype("float")
df["RH"] = df["RH"].str.replace(",",".").astype("float")
```

```python
df.head()
```

```python
# df = df.rename(columns={"T":"Temperature"})
df = df.rename(columns={"AH":"Absolute Humidity"})
df = df.rename(columns={"RH":"Relative Humidity"})
```

```python
df.describe()
```

```python
df['Absolute Humidity'] = df['Absolute Humidity'].multiply(100)
```

OUTLIERS REMOVAL

```python
def remove_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5 * IQR
    outliers_mask = (col > Q3 + threshold) | (col < Q1 - threshold)
    return col[~outliers_mask]
```

```python
df.columns
```

```python
remove_outliers(df['CO(GT)'])
```

```python
columns = ['CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)',
       'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)',
       'PT08.S5(O3)', 'T', 'Relative Humidity', 'Absolute Humidity']
```

```python
for i in columns:
    df[i] = remove_outliers(df[i])
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = df['CO(GT)'])
plt.title('Carbon-Di-Oxide')
plt.show()
```

```python
num_cols = len(columns)
rows = (num_cols + 2) // 3  # 3 plots per row

plt.figure(figsize=(15, 5*rows))

for idx, col in enumerate(columns):
    plt.subplot(rows, 3, idx + 1)
    sns.boxplot(data=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
sns.set(rc={'figure.figsize': (10, 10)})

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Clean 3D scatter plot — no palette, just basic color
ax.scatter(df['T'], df['Absolute Humidity'], df['Relative Humidity'], 
           c=df['T'], cmap='inferno', marker='.')

# Label it up
ax.set_xlabel('Temperature (°C)')
ax.set_ylabel('Absolute Humidity (g/m³)')
ax.set_zlabel('Relative Humidity (%)')

plt.title('3D Scatter: Temp vs Absolute & Relative Humidity')
plt.show()
```

```python
df_1 = df[['Date','T']].loc[:200]

df_1['Date'] = pd.to_datetime(df_1['Date'], dayfirst=True)

# Group by date, get mean temp
daily_avg = df_1.groupby('Date')['T'].mean().reset_index()
daily_avg = daily_avg.sort_values(by='T',ascending=True)
# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x='Date', y='T',hue='Date' ,data=daily_avg, palette='Pastel2')

plt.title('Average Temperature by Date')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
Humidity_data = df[['Date','Absolute Humidity','Relative Humidity']].loc[:200]
```

```python
# Make sure Date is in datetime format
Humidity_data['Date'] = pd.to_datetime(Humidity_data['Date'], dayfirst=True)

# Optional: group by Date if there are repeated timestamps
daily_avg = Humidity_data.groupby('Date')[['Absolute Humidity', 'Relative Humidity']].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))

plt.plot(daily_avg['Date'], daily_avg['Relative Humidity'], 
         label='Relative Humidity (%)', color='dodgerblue', marker='o')

plt.plot(daily_avg['Date'], daily_avg['Absolute Humidity'], 
         label='Absolute Humidity (g/m³)', color='darkorange', marker='s')

plt.title('Relative vs Absolute Humidity (First 200 Records)')
plt.xlabel('Date')
plt.ylabel('Humidity')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
