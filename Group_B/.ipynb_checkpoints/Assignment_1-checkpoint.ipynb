```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("dataset_Facebook.csv", sep=";")
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
post_engagement = df[[
    'Type',
    'Page total likes',
    'comment',
    'like',
    'share',
    'Total Interactions',
    'Post Month']]

post_reach = df[[
    'Type',
    'Lifetime Post Total Reach',                                              
    'Lifetime Post Total Impressions',
    'Post Month'
]]
```

```python
post_engagement
```

```python
post_reach
```

```python
merged_df = pd.merge(post_engagement, post_reach, on='Post Month', how='inner')
merged_df
```

```python
sorted_df = merged_df.sort_values(by=['Total Interactions', 'Page total likes'], ascending=[False, False])
sorted_df
```

```python
df['Post Month'] = pd.to_numeric(df['Post Month'], errors='coerce')
df['Total Interactions'] = pd.to_numeric(df['Total Interactions'], errors='coerce')

# correlation between 'Post Month' and 'Total Interactions'
correlation = df['Post Month'].corr(df['Total Interactions'])

print(f"Correlation between Post Month and Total Interactions: {correlation}")
```
