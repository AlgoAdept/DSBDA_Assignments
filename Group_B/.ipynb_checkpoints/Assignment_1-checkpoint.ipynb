```python
#Converting dictionaries into dataframes

Student_Details = {
    'Roll_no' : [10,20,30,40,50],
    'Name' : ['Aries','Jasper','Keifer','Calix','John']
}
Marks = {
    'Roll_no' : [10,20,30,40,50],
    'Internal_Score' : [50,46,37,45,27],
    'Written_Exam' : [75,48,69,80,70],
    'Total_Score' : [125,94,106,125,97]
}

Student_Details = pd.DataFrame(Student_Details)
Marks = pd.DataFrame(Marks)
```

```python
# DataFrames based on post type
photo_df = df[df['Type'] == 'Photo']
video_df = df[df['Type'] == 'Video']
status_df = df[df['Type'] == 'Status']
link_df = df[df['Type'] == 'Link']
```

```python
# DataFrame with interactions
interaction_df = df[['Type', 'like', 'share', 'comment']]
```

```python
# Merging the data
photo_merged = pd.merge(photo_df, interaction_df, on='Type')
video_merged = pd.merge(video_df, interaction_df, on='Type')
status_merged = pd.merge(status_df, interaction_df, on='Type')
link_merged = pd.merge(link_df, interaction_df, on='Type')
```

```python
# Grouping data by Type to analyze likes and comments
likes_comments_summary = interaction_df.groupby('Type')[['like', 'comment']].sum()
likes_comments_summary = likes_comments_summary.sort_values(by=['like','comment'], ascending = [False,False])
likes_comments_summary
```

```python
# Visualization
likes_comments_summary.plot(kind='bar', figsize=(10, 5), color=['purple', 'pink'])
plt.title('Comparison of Likes and Comments by Post Type')
plt.xlabel('Post Type')
plt.ylabel('Count')
plt.legend(['Likes', 'Comments'])
plt.show()
```
