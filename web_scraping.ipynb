```python
import requests 
import bs4 
```

```python
request1 = requests.get('https://www.meesho.com/stylish-casual-block-heels-for-women/p/5y67ap') 
request1 
```

```python
request1.content 
```

```python
soup = bs4.BeautifulSoup(request1.text)
```

Fetching Reviews / Comments

```python
reviews = soup.find_all('span',{'class' :
 'sc-eDvSVe gUjMRV Comment__CommentText-sc-1ju5q0e-3 cfdxfJ Comment__CommentText-sc-1ju5q0e-3 cfdxfJ'});
for review in reviews:
    print(review.get_text() + "\n")
```

Fetching ratings

```python
rating_tag = soup.find('h1', {'class': 'sc-eDvSVe cdZTwf'})
print("Overall Rating : ",rating_tag.get_text())
```

```python
Individual_ratings = soup.findAll('span',{'class':'sc-eDvSVe laVOtN'});
for indi_rating in Individual_ratings:
    print(indi_rating.get_text() + "\n")
```

Fetching Customer names

```python
cust_name = soup.findAll('span',{'class':"sc-eDvSVe dugLmN"})
for customer in cust_name:
    print(customer.get_text() + "\n")
```
